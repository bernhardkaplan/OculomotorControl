```python
# plot sigma_R in dependence of x_pre and v_stim
import numpy as np
import matplotlib
n_x = 20
n_v = 20
x_pre = np.linspace(0, 1., n_x)
v_stim = np.linspace(-2., 2., n_v)
# composite = np.abs(x_pre - .5) * np.abs(v_stim)
sigma_R = np.zeros((n_x, n_v))

a = 1.
b = 10.
c = .1
d = 100.
tau = 10.
w_x = 1.
w_v = 1.

for i_x in xrange(n_x):
    for i_v in xrange(n_v):
        x = w_x * np.abs(x_pre[i_x] - .5)**2 * w_v * np.abs(v_stim[i_v])
        sigma_R[i_x, i_v] = a / (b + d * np.exp(-tau * (x - c)))
    
    
norm_x = matplotlib.colors.Normalize(vmin=np.min(x_pre), vmax=np.max(x_pre))
m = matplotlib.cm.ScalarMappable(norm=norm_x, cmap=matplotlib.cm.jet)
m.set_array(x_pre)
rgba_colors_x = m.to_rgba(x_pre)

norm_v = matplotlib.colors.Normalize(vmin=np.min(v_stim), vmax=np.max(v_stim))
m = matplotlib.cm.ScalarMappable(norm=norm_v, cmap=matplotlib.cm.jet)
m.set_array(v_stim)
rgba_colors_v = m.to_rgba(v_stim)


fig = pylab.figure()
ax1 = fig.add_subplot(211)
ax2 = fig.add_subplot(212)
pylab.rcParams.update({'figure.subplot.vspace' : 0.5})
for i_x in xrange(n_x):
    ax1.plot(v_stim, sigma_R[i_x, :], label = '$x_{pre}=%.1f$' % x_pre[i_x], color=rgba_colors_x[i_x])     
             
for i_v in xrange(n_v):
    ax2.plot(x_pre, sigma_R[:, i_v], label = '$v_{stim} = %.1f$' % (v_stim[i_v]), color=rgba_colors_v[i_v])
    print v_stim[i_v]

fs=20

ax1.set_xlabel('$v_{stim}$', fontsize=fs)
ax2.set_xlabel('$x_{pre}$', fontsize=fs)

#ax1.legend(loc=(ax1.get_xlim()[1] * 1., .5))
#ax2.legend(loc=(ax2.get_xlim()[1] * 1., .5))

```
